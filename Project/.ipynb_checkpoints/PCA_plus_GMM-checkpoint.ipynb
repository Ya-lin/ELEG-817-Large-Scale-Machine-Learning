### Import packages

```python
import pandas as pd
from tqdm import trange
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
```

### Load data

```python
X_corr = pd.read_csv("Data_Correct.csv",header=None)
X_incorr = pd.read_csv("Data_Incorrect.csv",header=None)
X_corr = X_corr.to_numpy()
X_incorr = X_incorr.to_numpy()
```

### Explore data

```python
print(X_corr.shape)
print(X_incorr.shape)
```

```python
# (117x90)x240, plot the first three dimensions of first repetition
xc_1 = X_corr[0,:]
xc_2 = X_corr[117,:]
xc_3 = X_corr[234,:]
xi_1 = X_incorr[0,:]
xi_2 = X_incorr[117,:]
xi_3 = X_incorr[234,:]
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(xc_1,label="m1")
plt.plot(xc_2,label="m2")
plt.plot(xc_3,label="m3")
plt.xlabel("Time")
plt.ylabel("Motion angle")
plt.title("Correct")
plt.subplot(1,2,2)
plt.plot(xi_1,label="m1")
plt.plot(xi_2,label="m2")
plt.plot(xi_3,label="m3")
plt.xlabel("Time")
plt.ylabel("Motion angle")
plt.title("Incorrect")
plt.show()
```

### Transform data to be in size (240x90)x117

```python
n_dim = 117
T = X_corr.shape[1]
R = int(X_corr.shape[0]/n_dim)
Corr_Data = X_corr[0:n_dim,:].T
for i in range(1,R):
    indv = X_corr[i*n_dim:(i+1)*n_dim,:].T
    Corr_Data = np.concatenate((Corr_Data,indv),axis=0)
print(Corr_Data.shape)
```

```python
Incorr_Data = X_incorr[0:n_dim,:].T
for i in range(1,R):
    indv = X_incorr[i*n_dim:(i+1)*n_dim,:].T
    Incorr_Data = np.concatenate((Incorr_Data,indv),axis=0)
print(Incorr_Data.shape)
```

```python
print(Corr_Data[0:240,:].shape)
print(Incorr_Data[0:240,:].shape)
```

```python
# plot the first repetition
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(Corr_Data[0:240,:])
plt.ylim([-1,1])
plt.xlabel("Time")
plt.ylabel("Corrrect motion angle")
plt.title("Each curve stands for one dimension of motion curve")
plt.subplot(1,2,2)
plt.plot(Incorr_Data[0:240,:])
plt.ylim([-1,1])
plt.xlabel("Time")
plt.ylabel("Incorrrect motion angle")
plt.title("Each curve stands for one dimension of motion curve")
plt.show()
```

### Perform PCA by sklearn

```python
pca = PCA(n_components=4)
Corr_4 = pca.fit_transform(Corr_Data)
print(Corr_4.shape)
```

```python
Incorr_4 = pca.transform(Incorr_Data)
print(Incorr_4.shape)
```

```python
# plot the first repetition
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(Corr_4[0:240,:])
plt.xlabel("Time")
plt.ylabel("Correct Motion angle-reduced")
plt.title("Each curve stands for one dimension of motion curve")
plt.subplot(1,2,2)
plt.plot(Incorr_4[0:240,:])
plt.xlabel("Time")
plt.ylabel("Incorrect Motion angle-reduced")
plt.title("Each curve stands for one dimension of motion curve")
plt.show()
```

```python
Corr_recovered = pca.inverse_transform(Corr_4)
Incorr_recovered = pca.inverse_transform(Incorr_4)
print(Corr_recovered.shape)
print(Incorr_recovered.shape)
```

```python
# plot the first repetition of origianl vs reconstructed
plt.figure(figsize = (12,10))
plt.subplot(2,2,1)
plt.plot(Corr_Data[0:240,:])
plt.xlabel("Time")
plt.ylabel("Correct Motion angle-original")
plt.title("Each curve stands for one dimension of motion curve")
plt.subplot(2,2,2)
plt.plot(Corr_recovered[0:240,:])
plt.xlabel("Time")
plt.ylabel("Correct Motion angle-reconstructed")
plt.title("Each curve stands for one dimension of motion curve")
plt.subplot(2,2,3)
plt.plot(Incorr_Data[0:240,:])
plt.xlabel("Time")
plt.ylabel("Incorrect Motion angle-original")
plt.title("Each curve stands for one dimension of motion curve")
plt.subplot(2,2,4)
plt.plot(Incorr_recovered[0:240,:])
plt.xlabel("Time")
plt.ylabel("Incorrect Motion angle-reconstructed")
plt.title("Each curve stands for one dimension of motion curve")
plt.show()
```

### Train a GMM by sklearn

```python
gmm = GaussianMixture(n_components=6,n_init=10,init_params='kmeans')
```

```python
gmm.fit(Corr_4)
```

```python
print("weights: ", gmm.weights_)
print("means: ", gmm.means_)
print("covariances: ", gmm.covariances_)
```

```python
print("Converge or not: ", gmm.converged_)
print("Maximal iteration: ", gmm.n_iter_)
```

### Calculate negative likelihood

```python
X_corr4 = np.zeros((R,T,4))
X_incorr4 = np.zeros((R,T,4))
for r in range(R):
    X_corr4[r,:,:] = Corr_4[r*T:(r+1)*T,:]
    X_incorr4[r,:,:] = Incorr_4[r*T:(r+1)*T,:]
print(X_corr4.shape)
print(X_incorr4.shape)
```

```python
# average(over time) log likelihood, instead of average negative log likelihood
print(gmm.score(X_corr4[0,:,:]))
print(gmm.score(X_incorr4[0,:,:]))
```

```python
lik = {"corr":[],"incorr":[]}
for r in range(R):
    lik["corr"].append(gmm.score(X_corr4[r,:,:]))
    lik["incorr"].append(gmm.score(X_incorr4[r,:,:]))
```

```python
plt.scatter(np.arange(1,91),lik["corr"], label = "Correct Sequences")
plt.scatter(np.arange(1,91),lik["incorr"], label = "Incorrect Sequences")
plt.xlabel("Sequence Number")
plt.ylabel("Log-Likelihood")
plt.legend()
plt.show()
```
